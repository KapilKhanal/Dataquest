```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

# This will need to be done in the future so 
# get accustomed to using now
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
titanic = pd.read_csv('train.csv')
cols = ['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
titanic = titanic[cols].dropna()
```

```python
import seaborn as sns    # seaborn is commonly imported as `sns`
sns.distplot(titanic["Age"])
plt.show()
```

```python
sns.kdeplot(titanic["Age"], shade=True)
plt.xlabel("Age")
plt.show()
```

```python
sns.set_style('white')
sns.kdeplot(titanic['Age'], shade=True)
sns.despine(left=True, bottom=True)
plt.xlabel('Age')
```

```python
# size is deprecated, use height instead
g = sns.FacetGrid(titanic, col="Pclass", height=6)
g.map(sns.kdeplot, "Age", shade=True)
sns.despine(left=True, bottom=True)
plt.show()
```

```python
g = sns.FacetGrid(titanic, col="Survived", row="Pclass", hue="Sex", height=3)
g.map(sns.kdeplot, "Age", shade=True)
sns.despine(left=True, bottom=True)
plt.show()
```
